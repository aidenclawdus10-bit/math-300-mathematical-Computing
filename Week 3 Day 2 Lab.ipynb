### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
```

```python
def f(x):
    return x ** 2
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 100)
c = np.linspace(-1, 1, 1000)
fig, axes =plt.subplots(1,3, sharey=True)
axes[0].scatter(a, f(a), label='10 slices')
axes[1].scatter(b, f(b), label='100 slices')
axes[2].scatter(c, f(c), label='1000 slices')
axes[0].legend()
axes[1].legend()
axes[2].legend()
```

### Problem 2
Pick a function $f(x)$ you like and plot some values using `plt.scatter()`. Make sure your plot is as informative as possible.

```python
def f(x):
    return np.sin(x) * np.cos(x)
A = np.linspace(-np.pi, np.pi, 1000)
plt.scatter(A, f(A), label='cos(x) * sin(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```
