```python
df =pd.read_csv(r"C:\Users\aiden\Downloads\iris (1).csv")
plt.scatter(df["sepal.length"], df["sepal.width"])
plt.xlabel("length")
plt.ylabel("width")
plt.figure(figsize =(2,2))
Q1 = df["sepal.length"].quantile(0.25)
Q3 = df["sepal.width"].quantile(0.75)

IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df["sepal.length"] < lower) | (df["sepal.width"] > upper)]
if len(outliers)>1:
    print("there are outliers")
x = df["sepal.length"]
y = df["sepal.width"]
m, b = np.polyfit(x, y, 1)
y_pred = m * x + b
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r2 = 1 - ss_res / ss_tot
plt.figure(figsize=(6,6))
plt.scatter(x, y, label="Data")
plt.plot(x, y_pred, color="red", label="Regression Line")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.grid(True)
plt.show()
print(f"Slope = {m:.4f}")
print(f"Intercept = {b:.4f}")
print(f"R² = {r2:.4f}")
print(f"Equation: y = {m:.4f}x + {b:.4f}")
if r2>0.95:
    print("there seems to be a corrlation")
else:
    print("no corrlation")
plt.hist(df["variety"])
print(df["variety"].mode())
```

```python

df = pd.read_csv("DolphinNetwork.csv")
graph = {}
for _, row in df.iterrows():
    a = row["Source"]
    b = row["Target"]

    if a not in graph:
        graph[a] = set()

    if b not in graph:
        graph[b] = set()

    graph[a].add(b)
    graph[b].add(a)     
n = len(graph)
radius = 10
positions = {}
for i, node in enumerate(graph):
    angle = 2 * math.pi * i / n
    positions[node] = (radius * math.cos(angle),radius * math.sin(angle))

plt.figure(figsize=(12,12))
for node in graph:
    for neighbor in graph[node]:
        x1, y1 = positions[node]
        x2, y2 = positions[neighbor]

        plt.plot([x1, x2], [y1, y2],
                 color="gray", linewidth=1)
for node in graph:
    x, y = positions[node]
    plt.scatter(x, y,
                color="dodgerblue",
                s=120)
    plt.text(x, y, node,
             fontsize=8,
             ha="center",
             va="center")

plt.title("Dolphin Network")
plt.axis("off")
plt.show()
triangle_counts = {}
for node in graph:

    neighbors = list(graph[node])
    count = 0

    for i in range(len(neighbors)):
        for j in range(i + 1, len(neighbors)):

            if neighbors[j] in graph[neighbors[i]]:
                count += 1

    triangle_counts[node] = count
plt.figure(figsize=(15,6))
plt.bar(
    triangle_counts.keys(),
    triangle_counts.values(),
    color="steelblue")
plt.xticks(rotation=90)
plt.xlabel("Dolphin")
plt.ylabel("Triangle Count")
plt.title("Triangle Count per Dolphin")
plt.tight_layout()
plt.show()
highest = max(triangle_counts, key=triangle_counts.get)
lowest = min(triangle_counts, key=triangle_counts.get)
print("Highest:")
print(highest, triangle_counts[highest])
print()
print("Lowest:")
print(lowest, triangle_counts[lowest])
mean = sum(triangle_counts.values()) / len(triangle_counts)
print()
print("Mean triangle count:", mean)
```
