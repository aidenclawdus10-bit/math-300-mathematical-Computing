### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
import math as math 
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
````

````python
x = np.linspace(-5,5,100)
result = x**2 + 6*x +10 
plt.figure(figsize=(8,4))
plt.plot(x,result)
plt.xlabel("imput")
plt.ylabel("output")
from scipy.optimize import minimize
def result(x):
    return(x**2 + 6*x +10 )
minn = minimize(result, x0 = -2)
print(minn)
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
def f(y):
    return -50 *y + y**2
max_ = minimize(f,x0=25)
print(max_)
````

## Problem 3

Minimize

```python
np.sin(x)+x**2
```

using different starting values.

Do you obtain the same solution? Compute the derivative using `sympy`. When is this derivative zero?

````python
def f(x):
    return np.sin(x)+x**2
min_1 = minimize(f,x0=math.pi)
min_2 = minimize(f,x0=1/2*math.pi)
min_3 = minimize(f,x0=0)
min_4 = minimize(f,x0=10)
print(min_1)
print(min_2)
print(min_3)
print(min_4)
# i got the same min imput for each 
x = sy.symbols("x")
print(sy.diff(sy.sin(x)+x**2))
````
